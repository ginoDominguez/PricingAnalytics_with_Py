Definir el dataframe para la regresion:

```python
## Adecuar el dataset de manera que la demanda, el precio y costo del producto marca blanca sea comparable con el Producto prime
# variables a utilizar:
# Agrupadores: Fecha, Departamento, dia de la semana, mes del anio, clases
# Variables numericas: descuento, precio, costo, VentaKilos, Venta_Soles, CostoSoles, Rentabilidad

# Defino el dataset Total:
data_Total = data.groupby(['Fecha', 'Departamento' ,'dia_de_la_semana', 'mes_del_anio', 'semana_del_anio', 'dia_del_año','Clases' ]).agg(
    {
        'Venta_Kilos': 'sum',
        'Venta_Soles': 'sum',
        'Costo_Soles': 'sum',
        'Precio': np.average,
        'Costo': np.average,
        'Rentabilidad': np.average
    }
).reset_index()
## defino el dataste solo de Marca Blancam == Marca Rojo
data_MarcaBlanca= data[data['Producto'] == "Marca Rojo"]
data_MarcaBlanca= data_MarcaBlanca.groupby(['Fecha', 'Departamento' ,'Dscto', 'DsctoMarcaBlanca' ]).agg(
    {
        'Venta_Kilos': 'sum',
        'Venta_Soles': 'sum',
        'Costo_Soles': 'sum',
        'Precio': np.average,
        'Costo': np.average,
        'Rentabilidad': np.average
    }
).reset_index()
## Defino el datase solo marca Prime == Marca Azul
data_MarcaPrime= data[data['Producto'] == "Marca Azul"]
data_MarcaPrime= data_MarcaPrime.groupby(['Fecha', 'Departamento' ,'Dscto', 'DsctoMarcaBlanca' ]).agg(
    {
        'Venta_Kilos': 'sum',
        'Venta_Soles': 'sum',
        'Costo_Soles': 'sum',
        'Precio': np.average,
        'Costo': np.average,
        'Rentabilidad': np.average
    }
).reset_index()


```

```python
#Data para Kmeans:
#codificar la variable Departamento:
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
# LabelEncoder: codifica etiquetas de una variable categórica en valores numéricos entre 0 y el número de clases menos 1.
encoder = LabelEncoder().fit(data_ModeloRegresion['Departamento'].values)

data_ModeloRegresion['DepartamentoCodificado'] = encoder.transform(data_ModeloRegresion['Departamento'])
data_ModeloRegresion.info()
```

```python
##Normalizando las variables:
# Clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

# Agrupamos los datos a nivel De Departamento para la segmentación
data_ModeloRegresion_K = data_ModeloRegresion.groupby(['Departamento']).aggregate(
    {'Venta_KilosMBlanca':'sum', 'Venta_KilosMPrime':'sum', 'PrecioUnitMPrime':'mean', 'PrecioUnitMBlanca':'mean'}).reset_index()
data_ModeloRegresion_K.head()



#data_ModeloRegresion_K[['DepartamentoCodificado','Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']]= \
#  scaler.fit_transform(data_ModeloRegresion[['DepartamentoCodificado','Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']].to_numpy())
#data_ModeloRegresion_K.head()
```

```python
# Normalizamos las variables
scaler = StandardScaler()
data_ModeloRegresion_K[['Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']] = \
  scaler.fit_transform(data_ModeloRegresion_K[['Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']].to_numpy())
data_ModeloRegresion_K.head()
```

```python
data_ModeloRegresion_K.describe().round(2)
```

```python
# Método de codo para hallar el número de clusters óptimo
inertias = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data_ModeloRegresion_K.iloc[:, 1:5])
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Método del codo')
plt.xlabel('N° de clusters')
plt.ylabel('Inercia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
data_ModeloRegresion_K['segmento'] = kmeans.fit_predict(data_ModeloRegresion_K.iloc[:, 1:5])
data_ModeloRegresion_K.head()
```

```python
# Revertimos la estandarización
data_ModeloRegresion_K[['Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']] = \
  scaler.inverse_transform(data_ModeloRegresion_K[['Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']].round(6).to_numpy())

data_ModeloRegresion_K[['Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']] = \
  data_ModeloRegresion_K[['Venta_KilosMBlanca','Venta_KilosMPrime','PrecioUnitMPrime', 'PrecioUnitMBlanca']]###.astype(int)
```

```python
# Obtenemos los resultados por segmento
resultados = data_ModeloRegresion_K.groupby(['segmento']).aggregate(
    {'Departamento': 'count', 'Venta_KilosMBlanca': 'mean', 'Venta_KilosMPrime': 'mean', 'PrecioUnitMPrime':'mean', 'PrecioUnitMBlanca':'mean'}).round(0).reset_index()

# Resultados agregados de cada cluster
resultados
```
